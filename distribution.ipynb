```python
import re

# Initialize an empty list to store the numbers following "i="
i_values = []

# Open the file and read line by line
with open("output.txt", "r") as f:
    for line in f:
        # Use regular expression to find "i=" followed by numbers
        match = re.search(r'i\s*=\s*(\d+)', line)
        if match:
            # Append the number to the list
            i_values.append(int(match.group(1)))

# Now i_values contains all the numbers following "i="
print(i_values)

# If you'd like to compute some statistics
average = sum(i_values) / len(i_values)
print(f"Average: {average}")

maximum = max(i_values)
print(f"Maximum: {maximum}")

minimum = min(i_values)
print(f"Minimum: {minimum}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
plt.rcParams['font.sans-serif'] = ['Songti SC']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
# 拟合正态分布参数
mu, std = norm.fit(i_values)

# 绘制直方图
plt.hist(i_values, bins=20, density=True, alpha=0.6, color='g', edgecolor='black')

# 绘制拟合的正态分布曲线
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('i值')
plt.ylabel('频率')
plt.title(f'拟合结果：均值 = {mu:.2f}, 标准差 = {std:.2f}')
plt.figure(dpi=1200)
plt.show()
```

```python
import numpy as np

mean_value = np.mean(i_values)
median_value = np.median(i_values)
std_dev = np.std(i_values)

print(f"均值：{mean_value}")
print(f"中位数：{median_value}")
print(f"标准差：{std_dev}")
```

```python
i_values.sort()

n = len(i_values)

# 计算1%和99%处的数据值
lower_index = int(np.ceil(0.01 * n)) - 1  # Python的索引从0开始，所以减1
upper_index = int(np.ceil(0.99 * n)) - 1  # 同上

lower_value = i_values[lower_index]
upper_value = i_values[upper_index]

print(f"99% CDF 区间是 [{lower_value}, {upper_value}]")
```
